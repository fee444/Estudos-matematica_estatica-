```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from pandas_datareader import wb
```

ESTRUTURA DE DADOS PARA CRIAÇÃO DE VETORES EM PY 

```python
""""
MUDANDO O TAMANHO (SHAPE) DO VETOR
"""

vetor_1 = np.array([1,2,3])
type(vetor_1)
vetor_1.shape

vetor_2 = np.array([[1,2,3]])
vetor_2.shape

vetor_3 = np.array([[1], [2], [3]])
vetor_3.shape
```

```python
#CRIANDO O GRÁFICO DO VETOR 

u = [-2,3]

plt.figure(figsize = (5,5))

vetor = plt.arrow(0,0, u[0], u[1], head_width = .3, width = .1, color ='blue', length_includes_head = True)

plt.plot(0, 0, 'ko', markerfacecolor='yellow', markersize=8)

plt.grid(linestyle = '--', linewidth = .6 )

plt.axis('square')

plt.axis([-4,4,-4,4])

plt.legend([vetor], ['u'])

plt.show()
```

OPERAÇÕES ENTRE VETORES


```python
u = np.array([1,2])
v = np.array([4, -6])
```

```python
soma = u + v 
print(soma)
```
